## Exploratory Analysis and Visualization

In this section we will then start doing some exploratory data analysis of these solar power plants in India. Starting from the *Plant_1* dataframe we could start for instance to compute the mean for the *DAILY_YIELD* and the sum of the *TOTAL_YIELD*. 

```python
Plant_1_daily_yield_mean = Plant_1['DAILY_YIELD'].mean()
Plant_1_daily_yield_mean
```

```python
print('The average daily yield of plant 1 at this Indian solar power energy generation is: {:.3f} kWh'.format(round(Plant_1_daily_yield_mean, 3)))
```

```python
Plant_1_Total_daily_yield = Plant_1['DAILY_YIELD'].sum()
Plant_1_Total_daily_yield
print('The total daily yield of plant 1 at this Indian solar power energy generation is: {:.3f} kWh'.format(round(Plant_1_Total_daily_yield, 3)))
```

```python
Plant_2_daily_yield_mean = Plant_2['DAILY_YIELD'].mean()
print('The average daily yield of plant 2 at this Indian solar power energy generation is: {:.3f} kWh'.format(Plant_2_daily_yield_mean))
```

```python
Plant_2_Total_daily_yield = Plant_2['DAILY_YIELD'].sum()
print('The total daily yield of plant 2 at this Indian solar power energy generation is: {:.3f} kWh'.format(Plant_2_Total_daily_yield))
```

### Question 3

This is related with the definition of median, which gives us a measure of values in a dataset cut through 50% of all of its inputs. The *AC_POWER* dataset contains more than 50% zero measures of *AC_POWER*, and therefore its median is zero. But the same does not hold for Plant 1. We can confirm these findings by drawing respective dataframes giving us the information about the zero values in mentioned columns and then calculate respective median values.

```python
Plant_2_df = Plant_2[Plant_2.AC_POWER == 0]
Plant_2_df
```

```python
Plant_2_df = Plant_2[Plant_2.AC_POWER >= 241]
Plant_2_df
```

As we can see in both dataframes drawn above, the number of values of *AC_POWER* equal to zero is **35662** in plant 2, while the number of values greater than 241 (this is the number representing the mean of *AC_POWER* values for this plant, as shown below) is **21851**. Therefore its median metric must be **zero**.

```python
Mean_AC_Power_plant_2 = Plant_2['AC_POWER'].mean()
Mean_AC_Power_plant_2
```

```python
Median_AC_Power_plant_2 = Plant_2['AC_POWER'].median()
Median_AC_Power_plant_2
```

```python
Mean_AC_Power_plant_1 = Plant_1['AC_POWER'].mean()
Mean_AC_Power_plant_1
```

```python
Median_AC_Power_plant_1 = Plant_1['AC_POWER'].median()
Median_AC_Power_plant_1
```

Checking by analysing also Plant 1´s respective dataframes for infering both median and mean values yields:

```python
Plant_1_df = Plant_1[Plant_1.AC_POWER == 0]
Plant_1_df
```

```python
Plant_1_df = Plant_1[Plant_1.AC_POWER >= 41]
Plant_1_df
```

As we can see from the dataframes above the number of values in Plant 1 which display an *AC_POWER* greater than its median value is **34412**, whereas the number of values displaying **zero** is **31951**. Hence, that is why the median for Plant 1 is a number close to 41, it is actually 41.494. We can see further that the first value greater than 41 is actually 60.685714, which is a number quite bigger than 41. Plant 2 therefore have a much smoother distribution of *AC_POWER* values smaller than its mean value, in contrast with Plant 1.
